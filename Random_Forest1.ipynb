```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Synthetic financial data (e.g., income, debt ratio)
np.random.seed(42)
X = np.random.rand(100, 2)
y = np.random.choice([0, 1], size=100)  # 0: Low risk, 1: High risk

# Train Random Forest classifier
rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(X, y)

# Plot feature importance
importances = rf.feature_importances_
features = ['Income', 'Debt Ratio']
plt.figure(figsize=(6, 4))
plt.bar(features, importances, color='teal')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Random Forest Feature Importance for Credit Scoring')
plt.show()
```
